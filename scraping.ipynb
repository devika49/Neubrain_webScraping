```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException

import pandas as pd
import time
from selenium.webdriver.chrome.options import Options

chrome_options = Options()
chrome_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36")
driver = webdriver.Chrome(service=service, options=chrome_options)

# Set up ChromeDriver path
driver_path = 'C:/Users/User/Downloads/chromedriver-win64/chromedriver-win64/chromedriver.exe'  # Update this path if necessary
service = Service(driver_path)
driver = webdriver.Chrome(service=service)

# URL to the page with the table
url = "https://eprplastic.cpcb.gov.in/#/plastic/home/main_dashboard"
driver.get(url)
```

```python
# Cell 3: Wait for the Dropdown to Load
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "ng-select-container")))
```

```python

# Cell 4: Click the Dropdown
print("Clicking dropdown...")
dropdown = driver.find_element(By.CLASS_NAME, "ng-select-container")
driver.execute_script("arguments[0].click();", dropdown)
time.sleep(3)  # Allow time for dropdown options to load

```

```python

   # Cell 5: Locate and Print Options
try:
    # Click the dropdown to ensure it is open
    print("Clicking the dropdown...")
    dropdown = driver.find_element(By.CLASS_NAME, "ng-dropdown-panel ng-select-bottom")
    driver.execute_script("arguments[0].click();", dropdown)  # Use JavaScript for reliability
    time.sleep(1)  # Allow time for the dropdown to open

    print("Waiting for dropdown panel...")
    dropdown_panel = WebDriverWait(driver, 20).until(
        EC.visibility_of_element_located((By.XPATH, "//div[contains(@class, 'ng-option ng-option-selected ng-option-marked']"))  # Use XPath for broader selection
    )

    print("Dropdown panel is visible, now waiting for options...")
    options = WebDriverWait(driver, 20).until(
        EC.visibility_of_all_elements_located((By.CLASS_NAME, "ng-option"))  # Adjust if necessary
    )

    print(f"Found {len(options)} options.")
    for option in options:
        print(f"Option text: '{option.text}'")  # Debugging print
        if option.text.strip() == "100":
            print("Selecting '100' entries...")
            option.click()
            break
    else:
        print("Option '100' not found.")

except TimeoutException:
    print("TimeoutException: Unable to locate the dropdown panel or options within the dropdown.")
    print("Current page source:")
    print(driver.page_source)  # Inspect this output for debugging
    driver.save_screenshot('debug_screenshot.png')  # Optional screenshot for further inspection
except NameError:
    print("NameError: TimeoutException is not defined. Ensure it is imported correctly.")

```

```python
# Cell 6: Select Option "100"
for option in options:
    if option.text.strip() == "100":
        print("Selecting '100' entries...")
        option.click()
        break
else:
    print("Option '100' not found.")
```
