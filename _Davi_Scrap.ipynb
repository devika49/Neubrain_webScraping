```python
# imports here
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.action_chains import ActionChains
import time
import os
import wget
import csv
from bs4 import BeautifulSoup

```

```python
# specify the path to chromedriver.exe (correctly initialized using Service)
chrome_driver_path = 'C:/Users/tamil/chromedriver-win64/chromedriver-win64/chromedriver.exe'
service = Service(executable_path=chrome_driver_path)
driver = webdriver.Chrome(service=service)

# open the webpage
driver.get("https://eprplastic.cpcb.gov.in/#/plastic/home/main_dashboard")
```

```python

# Locate the button using its visible text
button = driver.find_element(By.XPATH, "//button[text()='Table']")

# Use ActionChains to move to the button and click
actions = ActionChains(driver)
actions.move_to_element(button).click().perform()
```

```python
# Find the table element using XPath
table = driver.find_element(By.XPATH, "//table[@id='simple-table-with-pagination']")

# Get the HTML content of the table
html_data = table.get_attribute('outerHTML')

# Parse the HTML using BeautifulSoup
soup = BeautifulSoup(html_data, 'html.parser')

# Extract the table data
data = []
for row in soup.find_all('tr'):
    row_data = [cell.get_text(strip=True) for cell in row.find_all('td')]
    if row_data:
        data.append(row_data)

# Write the data to a CSV file
with open('table_data.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerows(data)

print("CSV file saved: table_data.csv")
```

```python
# Locate the button by its tooltip text "Next"
button = driver.find_element(By.XPATH, "//button[@ngbtooltip='Next']")

# Use ActionChains to move to the button and click
actions = ActionChains(driver)
actions.move_to_element(button).click().perform()
```

```python
# Function to extract table data and append it to CSV
def extract_table_data(csv_filename):
    # Find the table element
    table = driver.find_element(By.XPATH, "//table[@id='simple-table-with-pagination']")
    # Get the HTML content of the table
    html_data = table.get_attribute('outerHTML')
    # Parse the HTML using BeautifulSoup
    soup = BeautifulSoup(html_data, 'html.parser')
    
    # Extract the table data
    data = []
    for row in soup.find_all('tr'):
        row_data = [cell.get_text(strip=True) for cell in row.find_all('td')]
        if row_data:
            data.append(row_data)

    # Append the data to a CSV file
    with open(csv_filename, 'a', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerows(data)

# Function to click the "Next" button
def click_next_button():
    # Locate the button by its tooltip text "Next"
    button = driver.find_element(By.XPATH, "//button[@ngbtooltip='Next']")
    # Use ActionChains to move to the button and click
    actions = ActionChains(driver)
    actions.move_to_element(button).click().perform()
```

```python
# Main loop to extract data and click "Next" until the button is disabled
csv_filename = 'table_data2.csv'
with open(csv_filename, 'w', newline='', encoding='utf-8') as csvfile:  # Initialize CSV file with header
    pass

while True:
    # Extract and save the table data
    extract_table_data(csv_filename)
    print("Data extracted and saved to CSV.")

    try:
        # Check if the "Next" button is disabled (assuming it has 'disabled' attribute when inactive)
        next_button = driver.find_element(By.XPATH, "//button[@ngbtooltip='Next']")
        if 'disabled' in next_button.get_attribute('class'):
            print("Next button is disabled. Exiting loop.")
            break
        
        # Click the "Next" button if not disabled
        click_next_button()
        time.sleep(2)  # Add delay to wait for the new table data to load
    except Exception as e:
        print(f"Exception occurred: {e}")
        break

# Close the driver
driver.quit()
```
